```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#畫函數圖形
畫出愛心


```python
# 繪製心形圖形
t = np.linspace(0, 2 * np.pi, 1000)  # 生成 0 到 2π 的數值序列
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

plt.figure(figsize=(6, 6))  # 調整圖形大小
plt.plot(x, y, color='red')  # 使用紅色繪製心形線
plt.axis('equal')  # 設定 x 和 y 軸的比例相同，確保心形看起來對稱
plt.axis('off') # 隱藏坐標軸
plt.title("Heart Shape")  # 設定圖形標題
plt.show()
```
